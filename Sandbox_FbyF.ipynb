```python
import pandas as pd
from datetime import datetime, timedelta

# Example DataFrame
data = {
    "abstime": [363078330, 363118185, 363158823, 3065363157],
    "time": [40191, 80046, 120684, 2702367560],
    "channel": [0, 0, 0, 0],
    "type": [101, 101, 101, 101],
    "location": ["C001", "C001", "C001", "C001"],
    "data1": [0.0, 0.0, 0.0, 0.0],
}

df = pd.DataFrame(data)
print(df)
# Correct experiment start timestamp
correct_start_time = datetime.strptime("2024-07-11 16:25:58", "%Y-%m-%d %H:%M:%S")

# Define ZT0 (e.g., 9:00 AM on the start date)
zt0_time = datetime.strptime("09:00:00", "%H:%M:%S").time()
zt0 = datetime.combine(correct_start_time.date(), zt0_time)

# Transform DataFrame: Create new columns
def transform_time(row):
    # Convert `time` from microseconds to seconds
    time_in_seconds = row["time"] / 1_000_000
    # Calculate full timestamp (`fullts`)
    full_timestamp = correct_start_time + timedelta(seconds=time_in_seconds)
    # Calculate Zeitgeber hours (`zhrs`)
    zhrs = (full_timestamp - zt0).total_seconds() / 3600
    return pd.Series([full_timestamp, zhrs, time_in_seconds])

df[["fullts", "zhrs", "exsecs"]] = df.apply(transform_time, axis=1)

# Drop old columns (optional)
df.drop(columns=["abstime", "time"], inplace=True)

# Display the updated DataFrame
df
```
